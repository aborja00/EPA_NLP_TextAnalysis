```python
import requests
from bs4 import BeautifulSoup
import os
import wget
import time 
from random import randrange 
```

```python
## EPA superfund site (proposed and new NLPs)
url = 'https://www.epa.gov/superfund/current-npl-updates-new-proposed-npl-sites-and-new-npl-sites'
response = requests.get(url)
response ##successful response 
```

```python
## Pulling all the HTML from the website
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
## printing it in a readable manner
print(soup.prettify())
```

```python
## Getting all pdf links off of the site 
# notes: when you do href= lambda what it is actually saying is <a href="document1.pdf">, href is "document1.pdf" → 'pdf' in href is True.

links = soup.find_all('a', href=lambda href: href and 'pdf' in href)
links
```

```python
## pulling all semspub links which contain NLP site data
saved_links= []
for l in links:
        if 'semspub' in l.attrs.get('href', ''):
            saved_links.append(l.attrs['href'])
```

```python
saved_links
```

```python
# 10 links, 2 for each site
len (saved_links)
```

```python
## bringing in my snoozer function
def sleep(start_time,end_time):
    '''
    Snoozer for scraping site
    '''
    snoozer= randrange(start_time,end_time)
    print(f"Snoozing for {snoozer} seconds")
    time.sleep(snoozer)
```

```python
link_count = 0
start_range, end_range = 15, 28 ## random seconds between scrapes
os.makedirs("EPA Links", exist_ok=True) #folder on my computer that I want all downloaded data to land in

for link in saved_links: 
    link_count += 1
    print(f"Downloading link {link_count} of {len(saved_links)}: {link}") ##keeping track of how many links are being downloaded 
    sleep(start_range, end_range)
    
    try:
        wget.download(link, out="EPA Links") ##download all links and put them in my folder in my computer
    except Exception as e:
        print(f"Error downloading {link}: {e}") ##in case there are any problem downloads
    
```

```python
import glob
import pandas as pd
```

```python
## pulling all files that I have downloaded into 

target_files = sorted(glob.glob("EPA Links/*.pdf"))
target_files
```

```python
#https://pypi.org/project/PyPDF2/
```

```python
from PyPDF2 import PdfReader
```

```python
## extracting text in from pdf files
for file in target_files:
    reader = PdfReader(file)
    for page in reader.pages:
        full_text += page.extract_text()
        
```

```python
## moving all information to a txt file 
with open("output_text.txt", "w", encoding="utf-8") as f:
    f.write(full_text)
```

```python
## What is in the txt file 
full_text
```

```python
# Regex
import re
```

```python
##1st Capturing Group (.*?) matches any character (except for line terminators)
site_names= re.findall(
    r"Name of Site:\s*(.*?)\n",
    full_text,
    re.DOTALL) 
site_names
```

```python
cleaned_sites= list(set(site_names))
print (cleaned_sites)
```

```python
## all epa id numbers have 3 characters and then a bunch of numbers. 

EPA_ID= re.findall(
    r"(EPA ID(?:#| No\.?)?:?\s*(\w{3}\d{9}))",
    full_text,
    re.DOTALL) 
```

```python
## There are five unique EPA IDS which match the amount of sites. Some are repeated more due to inconsitent files 
EPA_ID
```

```python
## they were all prepared in september... not as useful 
```

```python
## getting the regions the NLP sites are in. There are 10 different areas: https://www.epa.gov/aboutepa/regional-and-geographic-offices
EPA_region=re.findall(
    r"(EPA Region:?\s*\d+)",
    full_text,
    re.DOTALL)
EPA_region
```

```python
## all NLP sites are in different EPA regions
cleaned_regions = list(set(entry.strip().replace("EPA Region:", "EPA Region") for entry in EPA_region))
cleaned_regions
```

```python
## Scores for different sites. Sites receiving HRS scores of 28.50 and above are eligible for the NPL.
score= re.findall(
    r"HRS SITE SCORE\s*=?\s*(\d+\.\d+)",
    full_text,
    re.DOTALL)
score
```
